```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

```python
#Display the total number of players
total_number_players = len(purchase_data["SN"].unique())
Player_total = pd.DataFrame({"Total Players": [total_number_players]})
Player_total
```

```python
#Calculate number of unique items,
unique_items = len(purchase_data["Item ID"].unique())

#average price,
average_price = round(purchase_data["Price"].mean(), 2)

#number of purchases,
number_of_purchases = len(purchase_data["Purchase ID"])



#total revenue.
total_revenue = purchase_data["Price"].sum()

#Create summary table and display
Summary_total = pd.DataFrame({"Number of Unique Items": [unique_items],
                              "Average Price": [average_price], "Number of Purchases": [number_of_purchases],
                              "Total Revenue": [total_revenue]})

Summary_total["Average Price"] = Summary_total["Average Price"].astype(float).map("${:,.2f}".format)
Summary_total["Total Revenue"] = Summary_total["Total Revenue"].astype(float).map("${:,.2f}".format)
Summary_total
```

```python
# Count of Male, Female, and Other/ Non-Disclosed Players

Total_count_df = purchase_data.groupby("Gender").nunique()
total_count = Total_count_df[["SN"]]
Gender_summary_df = Total_count_df.rename(columns={"SN" : "Total Players"})

#Percentage of Other/ Non-Disclosed Players, Female Players,  Male Players

percentage_of_players = round((total_count / total_number_players)*100, 2).astype(str) + "%"
Gender_summary_df["Percentage of Players"] = percentage_of_players
Gender_player_summary_df = Gender_summary_df[["Total Players", "Percentage of Players"]]
Gender_player_summary_df
```

```python
#Obtain purchase count,  
Total_count_df

Gender_purchase_summary_df = Total_count_df[["Purchase ID"]]
Gender_purchase_summary_df = Gender_purchase_summary_df.rename(columns={"Purchase ID" : "Purchase Count"})

Gender_purchase_summary_df

#avg purchase price,
average_purchase_price = purchase_data.groupby("Gender")["Price"].mean()
Gender_purchase_summary_df["Average Purchase Price"] = round(average_purchase_price,2).astype(float).map("${:,.2f}".format)
Gender_purchase_summary_df
#total purchase priceCount
total_purchase_price = purchase_data.groupby("Gender")["Price"].sum()
Gender_purchase_summary_df["Total Purchase Value"] = (total_purchase_price).astype(float).map("${:,.2f}".format)
Gender_purchase_summary_df

#ave total price / per person


#Gender_purchase_summary_df["Avg Total Purchase per Person"] = Gender_purchase_summary_df["Total Purchase Value"] / Gender_purchase_summary_df["Purchase Count"] 
 
Gender_purchase_summary_df
#create a summary data frame for results
#Gender_purchase_summary_df
#give dislayed data cleaner format
#display summary dataframe
```
